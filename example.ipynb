```python
import pandas as pd
```

```python
mdcfl = pd.read_csv('Result/mdcfl/FedAvg_wesad.csv')
ccfl = pd.read_csv('Result/ccfl/FedAvg_wesad.csv')

df = mdcfl['loss']
norm_mdcfl = df.copy()

df = ccfl['loss']
norm_ccfl = df.copy()

import matplotlib.pyplot as plt
plt.figure(figsize=(10,6))
plt.plot(norm_mdcfl, label='MDCFL', color='red', alpha=0.5)
plt.plot(norm_ccfl, label='CosineCFL', color='blue', alpha=0.5)
plt.xlabel('Sample Index')
plt.ylabel('Cost Value')
plt.legend()
plt.title('WESAD: Cost value Comparison between MDCFL and CosineCFL')
plt.show()
```

```python
mdcfl = pd.read_csv('Result/mdcfl/FedAvg_wesad_sil.csv')
ccfl = pd.read_csv('Result/ccfl/FedAvg_wesad_sil.csv')

df = mdcfl[['mahalanobis','mahalanobis_sil']]
norm_mdcfl = df.copy()

df = ccfl[['cosine','cosine_sil']]
norm_ccfl = df.copy()
import matplotlib.pyplot as plt
plt.figure(figsize=(10,6))
plt.plot(norm_mdcfl['mahalanobis_sil'], label='MDCFL', color='r', alpha=0.5)
plt.plot(norm_ccfl['cosine_sil'], label='CosineCFL', color='blue', alpha=0.5)
# plt.plot([max([x,y]) for x,y in zip(norm_mdcfl['mahalanobis_sil'], norm_ccfl['cosine_sil'])], label='Hybrid', color='red', alpha=0.5)
plt.xlabel('Sample Index')
plt.ylabel('Silhouette Score')
plt.legend()
plt.title('WESAD: Silhouette Scores Comparison between MDCFL and CosineCFL')
plt.show()
```

```python
plt.figure(figsize=(4,3))
plt.hist([1 if x>y else 0 for x,y in zip(norm_mdcfl['mahalanobis_sil'], norm_ccfl['cosine_sil'])], color='blue', alpha=0.5)
plt.xticks([0,1], ['CosineCFL better', 'MDCFL better'])
plt.xlabel('Which Method Yields Higher Silhouette Scores')
plt.ylabel('Frequency')
plt.title('WESAD:Comparison between Mahalanobis and Cosine Silhouette Scores')
plt.legend()
plt.show()
```
